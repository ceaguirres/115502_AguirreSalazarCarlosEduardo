```python
"""
Leer una matriz cuadrada A, obtener su matriz triangular superior asociada U
y determinar que fila de U contiene la mayor cantidad de unos
"""

A=[] 
U=[] 
con1=[]

#separo variable para las matrices

o=int(input('digite de que orden es su matriz cuadrada: ')) #ingreso matriz


for i in range(o): #itero las filas
      A.append([]) 
      for j in range (o):          #itero las columnas ya ingresando valor
         A[i].append(int(input(f'escriba el digito # {j+1} de la fila: ')))
          
U=A.copy()      #hago una copia de A para poder trabajarla 
for i in range(o): 
      for j in range (o):          #ciclo para llegara a cada elemento
         if i<=j:           #condicion para que este en la triangular superior
             U[i][j]=A[i][j]     #se deja igual si se cumple
         else:
             U[i][j]=0  #sino es cero

print('la matriz triangular U es la siguiente: ')
for i in range(o):           
    print(A[i], end=' ') #imprime los resultados
    print()  #para imprimir de manera mas entendible 
    

def cont(i): #definimos funcion para contar los unos
    con=0
    for j in range (o):
        if U[i][j]==1:
            con+=1
    return con
            

for i in range(o):      #se almacenan los resultados en una lista aparte
    con1.append(cont(i))

fil=0   
maximo= 0
for m in range(len(con1)): #de a lista aparte se saca el mayor valor 
    if con1[m]> maximo:
        maximo= con1[m]
        fil=m+1 #se alamacena el valor de la fila correspondiente al mayor
print('la fila con más unos es la:',fil) #se imprimen los resultados
    
```

```python
print('Por favor ingrese la lista dentro de [] y separe los elementos con \
comas. Por ejemplo:[1,2,3]')
lista = eval(input('Ingrese una lista:'))

listas = [] #Guarda la sublista más larga ordenada ascendentemente

for i in range(len(lista)):
    aux = i       #Variables auxiliares que ayudan en la identación de las 
    aux_2 = -1    #listas ascendentes

    for j in range(i, len(lista)-1): #Separa las sublistas de forma ascendentes
        if lista[j] <= lista[j+1]:
            aux_2 = j+1
        else:
            break
    listas.append(lista[aux:aux_2+1])
    
lista_mayor = listas[0]
for i in listas:                   #Separa la sublista ascendente más larga
    if len(i) >= len(lista_mayor):
        lista_mayor = i
    
if len(lista_mayor) == 0:
    print('No se halló una lista.')
else:
    print(f'La lista buscada es: {lista_mayor}')
```
