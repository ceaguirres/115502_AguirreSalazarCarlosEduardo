```python
#Leer un número entero positivo y determinar la suma de sus dígitos pares. 
#Porejemplo,en el número 124,los dígitos pares son el 2 y el 4 y su suma vale 6

x=int(input('Digite un entero positivo:'))

digitos=[]
pares=[]
suma=0

while x > 0:
    digitos.append(x%10)  #separa el numero uno por uno
    x=x//10

for i in digitos:
    if i % 2 == 0:         #calcula cual numero es par
        pares.append(i)

for a in range(len(pares)):
    suma += pares[a]   #la suma se guardara en la lista pares
    
print('La suma de los digitos pares es:', suma)      
```

```python
"""
Leer un número entero y 
determinar si la suma de sus dígitos es un número de Fibonacci.
"""
def suma_digitos(x):
    cont=0
    while x>0:
        residuo=x%10             
        x=x//10                  
        if residuo>0:             
            cont=cont+residuo
    return cont

def numero_fibonacci(x):
    num1=0
    num2=1
    resultado=False
    while(num1<=x):
        if (num1==x):
            resultado=True
        suma=num1+num2
        num1=num2
        num2=suma
    return resultado

while True:
    try:
        num=int(input("Escriba un número entero = "))
        break
    except ValueError:
        print("El caracter ingresado es incorrecto.")

cont=suma_digitos(num)
if cont<0:
    cont=cont*(-1)

resultado=numero_fibonacci(cont)
if  resultado:
    print("La suma de los dígitos en valor absoluto del número ingresado \
es",cont,".\nY el número",cont,"es de la serie fibonacci.")
else:
    print("La suma de los dígitos en valor absoluto es el número",cont,"\n\
Y no es de la serie fibonacci.")
```

```python
"""
Haga un programa que lea las coordenadas (x1,y1,r1) y (x2,y2,r2) 
que corresponden al centro y al radio de dos círculos. Lea un punto de 
coordenadas (a,b). Determina si (a,b) está contenido: 
    a) dentro del círculo 1; 
    b) dentro del círculo 2; 
    c) dentro de ambos círculos; 
    d) fuera de ambos círculos. 
Recordemos que el círculo es el lugar geométrico de los puntos del plano cuya 
distancia a otro punto fijo, llamado centro, es menor o igual que una cantidad 
constante, llamada radio.
"""
def punto_pertenece_a_dos_cir(x1,y1,r1,x2,y2,r2,a,b):
    #formula de la circunferencia (a-x)**2+(b-y)**2=r**2
    #centro en(x,y)
    per_cir=(a-x1)**2+(b-y1)**2 ; per_cir_2=(a-x2)**2+(b-y2)**2
    rad1=r1**2                      ; rad2=r2**2
    
    return per_cir,rad1,rad2,per_cir_2
while True:
    try:
        X1=float(input("Ingrese la coordenada X del primer circulo = "))
        Y1=float(input("Ingrese la coordenada Y del primer circulo = "))
        R1=float(input("Ingrese el radio del primer circulo = "))
        X2=float(input("Ingrese la coordenada X del segundo circulo = "))
        Y2=float(input("Ingrese la coordenada Y del segundo circulo = "))
        R2=float(input("Ingrese el radio del segundo circulo = "))
        A=float(input("Ingrese la coordenada X a calcular = "))
        B=float(input("Ingrese la coordenada Y a calcular = "))
        break
    except ValueError:
        print("El caracter ingresado es incorrecto. Inténtelo de nuevo.")

per_cir,rad1,rad2,per_cir_2=punto_pertenece_a_dos_cir(X1,Y1,R1,X2,Y2,R2,A,B)
if per_cir<=rad1 and per_cir<=rad2:
    print("El punto pertenece a las dos circulos.")
elif per_cir<=rad1:
    print("El punto pertenece al circulo 1.")
elif per_cir_2<=rad2:
    print("El punto pertenece al circulo 2.")
else:
    print("El punto no pertenece a ningun circulo.")
```

```python
"""
Escribir un programa que reciba una cadena de texto y reporte:
a) cuántas letras vocales en mayúscula se entraron;
b) cuántas letras con tilde se entraron (minúsculas y mayúsculas);
c) cuántos dígitos se entraron,
d) cuántos espacios se entraron;
e) cuántas palabras reservadas se entraron.
"""
import keyword

x=input('Escriba una cadena de texto:')

v=['A','E','I','O','U','Á','É','Í','Ó','Ú']
vt=['á','é','í','ó','ú','Á','É','Í','Ó','Ú']
d=['1','2','3','4','5','6','7','8','9','0']
e=(' ')

l=x.split()
#contadores
cant_vocales=0       
cant_vocales_tilde=0
cant_digitos=0
cant_espacios=0
cant_reservadas=0

for i in range(len(l)):
    aux=list(l[i])
#verifica si en x esta lo pedido y lo almacena en la  lista, y aumenta el 
#contador
    for j in range(len(aux)): 
        if aux[j] in v:
            cant_vocales+=1
        if aux[j] in vt:
            cant_vocales_tilde+=1
        if aux[j] in d:
            cant_digitos+=1
        if aux[j] in e:
            cant_espacios+=1
    if l[i] in keyword.kwlist:
        cant_reservadas+=1
           
print('a)Cantidad vocales en mayuscula=', cant_vocales)
print('b)Cantidad letras con tílde=', cant_vocales_tilde)
print('c)Cantidad de digitos=', cant_digitos)
print('d)Cantidad de espacios=', cant_espacios)
print('e)Cantidad de palabras reservadas=', cant_reservadas)
```

```python
"""
Hacer un programa que lea dos palabras (podrían estar en mayúscula o minúscula)
y determine cual está primero en el diccionario. 
El programa debe soportar letras con tilde, la ü y la ñ.
"""

minus=['a','á','b','c','d','e','é','f','g','h','i','í','j','k','l','m','n','ñ',
      'o','ó','p','q','r','s','t','u','ú','ü','v','w','x','y','z']
mayus=['A','Á','B','C','D','E','É','F','G','H','I','Í','J','K','L','M','N','Ñ',
      'O','Ó','P','Q','R','S','T','U','Ú','Ü','V','W','X','Y','Z']

x1=input('Digite la primera palabra:')
x2=input('Digite la segunda palabra:')

y=[]
z=[]

for i in x1:
    if i in minus:
        c=minus.index(i)    #busca la posicion de la letra en la cadena y la 
                            #agrega  a la lista y
        y.append(c)
    
    if i in mayus:
        a=mayus.index(i)    #busca la posicion de la letra en la cadena y la 
                            #agrega  a la lista y
        y.append(a)

for i in x2:
    if i in minus:
        r=minus.index(i)    #busca la posicion de la letra en la cadena y la 
                            #agrega  a la lista y
        z.append(r)
    
    if i in mayus:
        o=mayus.index(i)    #busca la posicion de la letra en la cadena y la 
                            #agrega  a la lista y
        z.append(o)
        
if y < z :
    print('Esta primero en el diccionario:', x1)
else:
    print('Esta primero en el diciconario:', x2)
```
